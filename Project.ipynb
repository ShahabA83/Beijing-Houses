```python
housing_dropped = housing.drop(['Unnamed: 0','Cid','id','url'], axis=1)
```

```python
housing_null = pd.DataFrame(housing_dropped.isna().sum())
```

```python
housing_no_missing = housing_dropped.copy()
```

```python
housing_no_missing.loc[housing_no_missing.DOM.isnull(), 'DOM'] = housing_no_missing.DOM.apply(lambda x : 1)
```

```python
housing_no_missing = housing_no_missing.drop(housing_no_missing[housing_no_missing.elevator.isna() == True].index)
```

```python
housing_no_missing = housing_no_missing.drop(housing_no_missing[housing_no_missing.subway.isna() == True].index)
```

```python
housing_no_missing.info()
```

```python
housing_no_missing.elevator = housing_no_missing.elevator.apply(lambda x :'has elevator' if x==1 else 'no elevator')
housing_no_missing.subway = housing_no_missing.subway.apply(lambda x :'has subway' if x==1 else 'no subway')
def shahab(x) :
    if x == 1 :
        return 'unknown'
    elif x == 2:
        return 'mixed'
    elif x == 3:
        return 'brick and wood'
    elif x == 4:
        return 'concrete'
    elif x == 5:
        return 'steel'
    elif x == 6:
        return 'steel-concrete composite'
housing_no_missing.buildingStructure = housing_no_missing.buildingStructure.apply(shahab)
def sh(x):
    if x == 1:
        return 'other'
    elif x == 2:
        return 'rough'
    elif x == 3:
        return 'Simplicity'
    elif x == 4:
        return 'hardcover'
housing_no_missing.renovationCondition = housing_no_missing.renovationCondition.apply(sh)

housing_no_missing.head()
```

```python
print('constructionTime : ',housing_no_missing.constructionTime.unique())
print('floor : ',housing_no_missing.floor.unique())
```

```python
housing_no_missing = housing_no_missing.drop(housing_no_missing[housing_no_missing.constructionTime == '未知'].index)
housing_no_missing.constructionTime = housing_no_missing.constructionTime.astype('int')

housing_no_missing.head()
```

```python
housing_no_missing.floor = housing_no_missing.floor.apply(lambda x : x[2:]).astype('int')
housing_no_missing.head()
```

```python
housing_no_missing.info()
```

```python
import numpy as np
#housing_no_outlier = None # To-Do
Q1 = np.quantile(housing_no_missing.totalPrice,0.25)
Q3 = np.quantile(housing_no_missing.totalPrice,0.75)
IQR = Q3 - Q1
housing_no_missing = housing_no_missing.drop(housing_no_missing[(housing_no_missing.totalPrice > Q3+(1.5*IQR)) | (housing_no_missing.totalPrice < Q1-(1.5*IQR))].index)
```

$$ distance = acos(sinφ_1*sinφ_2 + cosφ_1*cosφ_2 * cos(λ_2-λ_1)) * 6371.0088 $$
$$ φ: \text{latitude} $$ 
$$ λ: \text{longitude} $$ 

```python
from math import radians
import math
capital_Lng = radians(116.4074)
capital_Lat = radians(39.9042)

lst1 = housing_no_missing['Lat'].apply(lambda x: np.sin(radians(x))) * np.sin(capital_Lat)
lst2 = housing_no_missing['Lat'].apply(lambda x: np.cos(radians(x))) * np.cos(capital_Lat)
lst3 = housing_no_missing['Lng'].apply(lambda x: np.cos(capital_Lng - radians(x)))
housing_no_missing['distanceToCapital'] = (lst1.values + lst2.values * lst3.values)
housing_no_missing['distanceToCapital'] = housing_no_missing['distanceToCapital'].apply(lambda x : math.acos(x) * 6371.0088)
housing_no_missing.head()
```

```python
housing_no_missing['pricePerSquare'] = housing_no_missing.totalPrice / housing_no_missing.square * 1000

housing_no_missing.head()
```

```python
index = housing_no_missing.district.value_counts().index.sort_values()
x1 = housing_no_missing.groupby('district')['pricePerSquare'].mean()
x1.name = None
districts_df = pd.DataFrame(data = x1, index = index ,columns=['pricePerSquare mean'])

districts_df['% houses with elevator'] = housing_no_missing.groupby('district')['elevator'].agg(lambda x : (x=='has elevator').sum()*100 / x.count())
districts_df['% houses with subway'] = housing_no_missing.groupby('district')['subway'].agg(lambda x : (x=='has subway').sum()*100 / x.count())
districts_df['square mean'] = housing_no_missing.groupby('district')['square'].mean()
districts_df['buildeng age mean'] = housing_no_missing.groupby('district')['constructionTime'].agg(lambda x : 2021 - x.mean())
districts_df['frequent buildingStructure'] = housing_no_missing.groupby('district')['buildingStructure'].agg(lambda x : x.value_counts().head(1).index)
districts_df
```
